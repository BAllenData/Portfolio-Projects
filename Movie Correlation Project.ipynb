```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of hte plots we will create


#Read in the data

df = pd.read_csv(r'C:\Users\Big Dog Bryan\Downloads\movies.csv')
```

```python
# Checking out the data broadly

df.head()
```

```python
# Checking to see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#We have null values, and we want to convert our numbers from float to int, so we're going to remove null rows.

df = df.dropna(axis=0)

# Data types for our columns

df.dtypes
```

```python


# Changing column data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['released'] = df['released'].astype('str')
```

```python
df
```

```python
new = df['released'].str.partition(', ',True)
df['ReleaseMonth'] = new[0]
df['ReleaseSplit'] = new[2].str[:4]
df=df.drop(columns=['ReleaseMonth'])
df

```

```python
df.rename({'ReleaseSplit':'ReleaseYear'}, axis = "columns")

df
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)

#Looking for duplicates in the data.  Drop any if they exist.  
#The output of this function provides you the distinct count of values found in the specified column.

df['company'].drop_duplicates().sort_values(ascending = False)

df.drop_duplicates()
```

```python
#Hypothesis:  Budget will have high correlation to gross
#Hyp 2:  Company will have high correlation to gross

#Scatter plot with budget vs. gross revenue

#Using matplotlib pyplot

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earning')

plt.xlabel('budget')
plt.ylabel('gross earning')
plt.show()
```

```python
df.head()
```

```python
#Now plotting budget vs gross earnings using Seaborn library

sns.regplot(x = 'budget', y = 'gross',data=df, scatter_kws = {"color":"red"}, line_kws={"color":"blue"})
```

```python
#Now that we have a scatter, let's look at the correlation
df.corr() #Pearson method
```

```python
#Table shows a high correlation between budget and gross

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Looks at Company

df.head()
```

```python
#This creates a numeric representation of the non-numeric fields in your data frame so that you can run a correlation analysis on all of them.

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#Now that we've represnted every field with numeric values, let's re-run the correlation matrix using our new df_numerized

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
df_numerized.corr()
```

```python
#Unstacking

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
